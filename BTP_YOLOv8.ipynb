```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde

# Create a DataFrame
df = pd.DataFrame(bubble_diam, columns=["Bubble Diameter (mm)"])

# Compute histogram data for fitting
counts, bin_edges = np.histogram(df["Bubble Diameter (mm)"], density=True)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2  # Get the center of each bin

# Define fitting functions
def logarithmic(x, a, b):
    return a * np.log(x) + b

def semi_logarithmic(x, a, b):
    return a * np.log(x + 1) + b  # Adding 1 to avoid log(0) issues

def normal(x, a, mu, sigma):
    return a * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))

# Filter out invalid bins for log-based fitting (logarithm is undefined for x <= 0)
valid_idx = bin_centers > 0
bin_centers_valid = bin_centers[valid_idx]
counts_valid = counts[valid_idx]

# Fit models
log_params, _ = curve_fit(logarithmic, bin_centers_valid, counts_valid)
semi_log_params, _ = curve_fit(semi_logarithmic, bin_centers_valid, counts_valid)
normal_params, _ = curve_fit(normal, bin_centers, counts)

# Calculate predicted values
log_pred = logarithmic(bin_centers_valid, *log_params)
semi_log_pred = semi_logarithmic(bin_centers_valid, *semi_log_params)
normal_pred = normal(bin_centers, *normal_params)

# KDE (non-parametric Gaussian)
kde = gaussian_kde(df["Bubble Diameter (mm)"])
kde_values = kde(bin_centers)

# Calculate R2 scores
log_r2 = r2_score(counts_valid, log_pred)
semi_log_r2 = r2_score(counts_valid, semi_log_pred)
normal_r2 = r2_score(counts, normal_pred)
kde_r2 = r2_score(counts, kde_values)

# Print R2 scores
print(f"Logarithmic R2: {log_r2:.4f}")
print(f"Semi-Logarithmic R2: {semi_log_r2:.4f}")
print(f"Normal R2: {normal_r2:.4f}")
print(f"KDE R2: {kde_r2:.4f}")

# Plot data and fitted models
plt.figure(figsize=(10, 6))

# Histogram
plt.hist(df["Bubble Diameter (mm)"], density=True, alpha=0.5, color="gray", label="Histogram")

# Plot fits
plt.plot(bin_centers_valid, log_pred, label=f"Logarithmic Fit (R2: {log_r2:.4f})", color="blue")
plt.plot(bin_centers_valid, semi_log_pred, label=f"Semi-Logarithmic Fit (R2: {semi_log_r2:.4f})", color="green")
plt.plot(bin_centers, normal_pred, label=f"Normal Fit (R2: {normal_r2:.4f})", color="red")
plt.plot(bin_centers, kde_values, label=f"Gaussian KDE (R2: {kde_r2:.4f})", color="orange")

# Labels and legend
plt.xlabel("Bubble Diameter (mm)")
plt.ylabel("Density")
plt.title("Bubble Diameter Distribution with Model Fits")
plt.legend()
plt.show()
```

```python

from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_spheroid(a, b, is_prolate=True):
    """
    Plots a spheroid.

    Parameters:
    a -- semi-major axis
    b -- semi-minor axis
    is_prolate -- True for prolate spheroid, False for oblate spheroid
    """

    # Define the angles for parameterization
    u = np.linspace(0, 2 * np.pi, 100)  # Angle in xy-plane
    v = np.linspace(0, np.pi, 100)      # Angle for z

    # Create meshgrid
    u, v = np.meshgrid(u, v)

    # Parametric equation for spheroid
    if is_prolate:
        # Prolate spheroid: semi-major axis along z
        x = a * np.sin(v) * np.cos(u)
        y = a * np.sin(v) * np.sin(u)
        z = b * np.cos(v)
    else:
        # Oblate spheroid: semi-major axis along x and y
        x = b * np.sin(v) * np.cos(u)
        y = b * np.sin(v) * np.sin(u)
        z = a * np.cos(v)

    # Plotting the spheroid
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, color='b', alpha=0.7, rstride=5, cstride=5)

    # Labeling the axes
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')

    plt.show()


def spheroid_surface_area(a, b):
    if a > b:
        # Oblate spheroid (flattened)
        e = np.sqrt(1 - (b**2 / a**2))  # Eccentricity
        surface_area = 2 * np.pi * a**2 + (2 * np.pi * b**2 / e) * np.log((1 + e) / (1 - e))
    else:
        # Prolate spheroid (elongated)
        e = np.sqrt(1 - (a**2 / b**2))  # Eccentricity
        surface_area = 2 * np.pi * a**2 + (2 * np.pi * a * b / e) * np.log((1 + e) / (1 - e))

    return surface_area


for result in results:
    masks = result.masks  # masks (polygon points)
    annotated_frame = image.copy()

    # masks = bubbles
    for mask in masks:

        contour_points = mask.xy
        boundary = []

        for points in contour_points:
            for point in points:

                x, y = int(point[0]), int(point[1])

                boundary.append([x, y])

        boundary = np.array(boundary, dtype=np.float32)

        ellipse = cv2.fitEllipse(boundary)
        # print(ellipse)

        (x_c, y_c), (a, b), theta = ellipse

        diameter = ((a*a*b) ** (1/3))  # Effective diameter

        surf_area = spheroid_surface_area(a, b)

        print(f"Center: ({x_c}, {y_c}), Diameter: {diameter:.2f}, Angle: {theta:.2f}, Surface Area(spheroid): {surf_area:.2f}")

        cv2.ellipse(annotated_frame, ((x_c, y_c), (a, b), theta), (0, 255, 0), 2)


        # Plot oblate spheroid (flattened along z-axis)
        # plot_spheroid(a, b, is_prolate=False)


    cv2_imshow(annotated_frame)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

```

```python
# Iterate through the dataset's labels
import os
import numpy as np
import cv2

def transform_labels(label_dir):
    for label_file in os.listdir(label_dir):
        if label_file.endswith(".txt"):
            # Get corresponding image to retrieve dimensions
            image_height = 544
            image_width = 544  # Get actual image dimensions

            with open(os.path.join(label_dir, label_file), "r") as f:
                label_data = f.readlines()

            new_labels = []
            for line in label_data:
                parts = line.strip().split()
                class_id = parts[0]
                points = list(map(float, parts[1:]))
                points = np.array(points).reshape(-1, 2)  # Reshape into (x, y) pairs

                if len(points) >= 5:  # Ensure enough points for ellipse fitting
                    # Convert normalized points back to pixel coordinates
                    points[:, 0] *= image_width  # x coordinates
                    points[:, 1] *= image_height  # y coordinates

                    # Step 3: Fit an ellipse to these points
                    # Convert points to float32 before calling fitEllipse
                    points = points.astype(np.float32)
                    ellipse = cv2.fitEllipse(points)

                    # Step 4: Extract ellipse parameters
                    (x_c, y_c), (a, b), theta = ellipse  # Center, axes, and angle

                    # Normalize the parameters based on the image size
                    x_c_norm = x_c / image_width
                    y_c_norm = y_c / image_height
                    a_norm = a / image_width
                    b_norm = b / image_height
                    theta_norm = np.deg2rad(theta)  # Convert theta to radians (YOLO prefers radians)

                    # Create new label format (assuming class_id = 0 for bubble)
                    new_label = f"{class_id} {x_c_norm:.6f} {y_c_norm:.6f} {a_norm:.6f} {b_norm:.6f} {theta_norm:.6f}\n"
                    new_labels.append(new_label)

            # Step 5: Write new labels back to the file (overwrite or create new)
            with open(os.path.join(label_dir, label_file), "w") as f:
                f.writelines(new_labels)

label_dir = f"{dataset.location}/train/labels"
transform_labels(label_dir)

label_dir = f"{dataset.location}/test/labels"
transform_labels(label_dir)

label_dir = f"{dataset.location}/valid/labels"
transform_labels(label_dir)

```
